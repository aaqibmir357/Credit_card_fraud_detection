Importing the libaries pandas for dataframe, Function train_test_split from sklearn for splitting the data into training set and test set, LogisticRegression function for model as finding the fraud detection of credit card is just a matter of yes or no meaning 0 or 1, which is a binary data hence logistic model.More over function accuracy_score from sklearn for determing the accuracy of our model.

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading the data set using pandas 

```python
df = pd.read_csv('/content/creditcard.csv') 
```

checking the null values in data set

```python
df.isnull().sum()                           
```

cleaning the data set by droping the null values

```python
credit_card_data = df.dropna()  
```

printing the data set

```python
credit_card_data                     
```

Describing the data set

```python
credit_card_data.describe()
```

Getting the information of data set

```python
credit_card_data.info()
```

counting the total  1 values and 0 values in class column 

```python
credit_card_data['Class'].value_counts()
```

the number of zeros indicating the legal trasactions is quit large and number of ones indicating the fraud trasactions is small in the data set, so we need to create the new_data_set from the existing one for equal distribution of 0 and 1


```python
legal = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
legal
fraud
```

```python
print(legal.shape)
print(fraud.shape)
```

```python
credit_card_data.groupby('Class').mean()
```

above we have compared the mean transaction of both class


creating the new data set containg the equal distribution of 0 and 1

```python
new_legal = legal.sample(n=88)
```

now concatnate this legal transaction table with 88 zeros with the fraud table using pandas libary

we have got the new_data_set with legal transaction and fraud transaction equal in number

Concatinating the two tables named by new_legal and fraud

```python
new_data_set = pd.concat([new_legal, fraud])
new_data_set
```

Splitting new_data set into features and targets(0,1)

X and Y representing the features and targets respectively.

```python
X = new_data_set.drop(columns='Class', axis=1)
Y= new_data_set['Class']
```

Now we will split the data(X,Y) into training data and test data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size=0.4, stratify=Y, random_state = 2)

print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)

```

Training the model

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

finding the accuracy score of our model


Model evaluation based on accuarcy score;
accuracy on training data


```python
X_train_predict = model.predict(X_train)
train_accuracy = accuracy_score(X_train_predict, Y_train)
print("accuracy on training data is : {}".format(train_accuracy))
```

predicting the accuracy on test data

```python
X_test_predict = model.predict(X_test)
test_accuarcy = accuracy_score(X_test_predict, Y_test)
print("the test accuracy is : {:.3f}".format((test_accuarcy)))

```

since the accuracy score of training data set and test data set is almost the same hence our model works well.
